```python
import time
import matplotlib.pyplot as plt
import random

# Генерація масивів
n_values = [5, 10, 20, 50, 100, 500, 1000, 2000, 3000]

# Функція для вимірювання часу
def measure_time(sort_func, arr):
    start = time.time()
    sort_func(arr)
    return time.time() - start

# Реалізація алгоритму сортування методом бульбашки
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Реалізація алгоритму сортування методом вставки
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Списки для збереження часу виконання
bubble_times = []
insertion_times = []

# Для виведення часу на екран
for n in n_values:
    arr = [random.randint(1, 10000) for _ in range(n)]
    
    # Копії масивів
    arr_bubble = arr[:]
    arr_insertion = arr[:]
    
    # Час для Bubble Sort
    t_bubble = measure_time(bubble_sort, arr_bubble)
    bubble_times.append(t_bubble)
    
    # Час для Insertion Sort
    t_insertion = measure_time(insertion_sort, arr_insertion)
    insertion_times.append(t_insertion)

    # Виведення часу для кожного розміру масиву
    print(f"n = {n}: Час виконання (Bubble Sort) = {t_bubble:.6f} секунд, Час виконання (Insertion Sort) = {t_insertion:.6f} секунд")

# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(n_values, bubble_times, label="Bubble Sort", marker='o', color='blue')
plt.plot(n_values, insertion_times, label="Insertion Sort", marker='o', color='green')
plt.xlabel('Розмір масиву (n)')
plt.ylabel('Час виконання (сек)')
plt.title('Порівняння часу виконання алгоритмів')
plt.legend()
plt.grid(True)
plt.show()
```
