## Оцінка асимптотичної складності алгоритму бінарного пошуку

Бінарний пошук — це ефективний алгоритм пошуку, який працює тільки на відсортованих масивах. Алгоритм працює наступним чином:

1. Спочатку визначаємо середній елемент масиву.
2. Якщо середній елемент є шуканим значенням, то пошук завершено.
3. Якщо шукане значення більше за середній елемент, то шукаємо у правій половині масиву.
4. Якщо шукане значення менше за середній елемент, то шукаємо в лівій половині масиву.

При кожному порівнянні ми відкидаємо половину елементів, що залишилися для подальшого пошуку.

### Оцінка часу:

Кожен крок алгоритму зменшує кількість елементів для пошуку вдвічі. Тому для масиву розміру \( $ n $ \) кількість кроків буде визначатися за допомогою логарифма від \( $ n $ \), тобто:

- **Часова складність** бінарного пошуку: \( $ O(\log n) $ \).

### Просторова складність:

Бінарний пошук використовує тільки кілька змінних для збереження меж пошуку (ліву, праву та середину), тому його **просторова складність**: \( $ O(1) $ \) (якщо ми не враховуємо стек викликів в рекурсивному варіанті).

### Формально:

- **Часова складність**: \( $ O(\log n) $ \)
- **Просторова складність**: \( $ O(1) $ \)

### Висновок:

Бінарний пошук є дуже ефективним алгоритмом для пошуку елементів у відсортованих масивах, оскільки його складність значно менша, ніж у лінійного пошуку (який має складність \( $ O(n) $ \)).


### Завдання 2

```python
import numpy as np
import pandas as pd
import time

# Лінійний пошук
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Функція для вимірювання часу лінійного пошуку
def measure_linear_search_time(n):
    arr = np.random.randint(0, 100000, size=n)  # випадковий масив
    target = np.random.randint(0, 100000)  # випадкове значення для пошуку
    start_time = time.time()
    linear_search(arr, target)
    end_time = time.time()
    return end_time - start_time

# Створення датафрейму для лінійного пошуку
def create_linear_search_df(max_n):
    data = {'n': [], 'time': []}
    for n in range(1, max_n + 1):
        data['n'].append(n)
        data['time'].append(measure_linear_search_time(n))
    return pd.DataFrame(data)

# Приклад виклику функції
linear_search_df = create_linear_search_df(100)  # Оцінка для масивів до 100 елементів
print(linear_search_df)
```

## Оцінка часу лінійного пошуку

### 1. Лінійний пошук
Лінійний пошук — це алгоритм, який перевіряє елементи масиву один за одним, поки не знайде шуканий елемент або не завершить перевірку всіх елементів. Це простий, але не дуже ефективний алгоритм для великих масивів.

**Алгоритм:**
- Починаємо з першого елемента масиву.
- Перевіряємо кожен елемент масиву на рівність з шуканим значенням.
- Якщо знайдено, повертаємо індекс цього елемента.
- Якщо елемент не знайдено після перевірки всіх елементів, повертаємо значення `-1`, що означає відсутність елемента.

**Часова складність:**  
Часова складність лінійного пошуку — \ $ (O(n) $ \), де \ $ (n  \ $)   — це кількість елементів у масиві. У найгіршому випадку, коли шуканий елемент знаходиться в кінці масиву або його немає взагалі, доведеться перевірити всі елементи. Тому часова складність прямо пропорційна розміру масиву.

### 2. Оцінка часу виконання лінійного пошуку
Для оцінки часу виконання лінійного пошуку ми можемо виміряти час, який витрачається на пошук у масиві різних розмірів.

Для цього ми генеруємо масиви різного розміру та вимірюємо час, який алгоритм витрачає на пошук елемента в кожному масиві.

- Як тільки ми виміряємо час виконання для масивів різного розміру, можна створити таблицю або графік, щоб побачити, як змінюється час виконання в залежності від розміру масиву.

### 3. Залежність часу від розміру масиву
Результати вимірювання часу виконання лінійного пошуку для масивів різного розміру можуть бути представлені в таблиці або графіку. Час виконання буде зростати лінійно з ростом розміру масиву.

### 4. Висновки
- Лінійний пошук має складність \(O(n)\), що означає, що з часом його ефективність знижується при збільшенні розміру масиву.
- Для великих масивів, де швидкість виконання є важливою, лінійний пошук може стати неефективним.
- Однак для невеликих наборів даних або коли потрібно виконати пошук у неструктурованих даних, лінійний пошук може бути цілком прийнятним варіантом.


### Завдання 3

```python
import numpy as np
import pandas as pd
import time

# Бінарний пошук
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Функція для вимірювання часу бінарного пошуку
def measure_binary_search_time(n):
    arr = np.sort(np.random.randint(0, 100000, size=n))  # випадковий відсортований масив
    target = np.random.randint(0, 100000)  # випадкове значення для пошуку
    start_time = time.time()
    binary_search(arr, target)
    end_time = time.time()
    return end_time - start_time

# Створення датафрейму для бінарного пошуку
def create_binary_search_df(max_n):
    data = {'n': [], 'time': []}
    for n in range(1, max_n + 1):
        data['n'].append(n)
        data['time'].append(measure_binary_search_time(n))
    return pd.DataFrame(data)

# Приклад виклику функції
binary_search_df = create_binary_search_df(100)  # Оцінка для масивів до 100 елементів
print(binary_search_df)
```

## Оцінка часу бінарного пошуку

### 1. Алгоритм бінарного пошуку
Бінарний пошук — це ефективний алгоритм пошуку елемента в відсортованому масиві. Алгоритм працює за принципом ділення масиву на дві частини, перевіряючи середній елемент на рівність шуканому. Якщо елемент знайдено, повертається його індекс. Якщо елемент менший або більший за середній, пошук продовжується лише в тій половині масиву, де шуканий елемент може бути знайдений.

**Кроки алгоритму:**
1. Масив повинен бути відсортованим.
2. Ініціалізуємо два індекси: лівий (`left`) та правий (`right`).
3. Поки `left` менше або рівне `right`, обчислюємо середній індекс.
4. Якщо елемент на середньому індексі рівний шуканому, повертаємо його індекс.
5. Якщо шуканий елемент більший за середній, продовжуємо пошук в правій частині масиву.
6. Якщо шуканий елемент менший, шукаємо в лівій частині масиву.
7. Якщо елемент не знайдено, повертаємо `-1`.

### 2. Оцінка часу виконання бінарного пошуку
Щоб оцінити час виконання бінарного пошуку для масивів різного розміру, необхідно:
- Для кожного розміру масиву \( n \) згенерувати випадковий відсортований масив.
- Проводити пошук в цьому масиві для випадково вибраного елемента.
- Виміряти час, який знадобиться для виконання бінарного пошуку на кожному масиві.

### 3. Збір даних
Функція **measure_binary_search_time(n)** створює випадковий відсортований масив розміру \( n \) та вимірює час виконання бінарного пошуку на цьому масиві.

- Масиви створюються випадковим чином за допомогою функції `np.random.randint()`.
- Для кожного масиву розміру \( n \) ми обираємо випадкове значення для пошуку, що дозволяє оцінити час пошуку елемента.

### 4. Створення датафрейму
За допомогою функції **create_binary_search_df(max_n)** ми вимірюємо час виконання бінарного пошуку для масивів різного розміру (від 1 до \( $ max\_n $ \)) та формуємо таблицю з полями:
- `n`: розмір масиву,
- `time`: час виконання бінарного пошуку для масиву розміру \( n \).

### 5. Висновки
- Бінарний пошук є значно ефективнішим за лінійний пошук завдяки своїй логарифмічній складності \( $ O(\log n) $ \).
- Час виконання бінарного пошуку зростає значно повільніше, порівняно з лінійним пошуком, оскільки при кожному кроці алгоритм зменшує кількість елементів вдвічі.
- Оцінка часу виконання дозволяє порівняти ефективність бінарного пошуку з іншими алгоритмами пошуку, зокрема лінійним пошуком.


### Завдання 4

```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt

# Лінійний пошук
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Бінарний пошук
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Функція для вимірювання часу лінійного пошуку
def measure_linear_search_time(n):
    arr = np.random.randint(0, 100000, size=n)
    target = np.random.randint(0, 100000)
    start_time = time.time()
    linear_search(arr, target)
    end_time = time.time()
    return end_time - start_time

# Функція для вимірювання часу бінарного пошуку
def measure_binary_search_time(n):
    arr = np.sort(np.random.randint(0, 100000, size=n))  # Відсортований масив
    target = np.random.randint(0, 100000)
    start_time = time.time()
    binary_search(arr, target)
    end_time = time.time()
    return end_time - start_time

# Створення датафрейму для лінійного пошуку
def create_linear_search_df(max_n):
    data = {'n': [], 'time': []}
    for n in range(1, max_n + 1):
        data['n'].append(n)
        data['time'].append(measure_linear_search_time(n))
    return pd.DataFrame(data)

# Створення датафрейму для бінарного пошуку
def create_binary_search_df(max_n):
    data = {'n': [], 'time': []}
    for n in range(1, max_n + 1):
        data['n'].append(n)
        data['time'].append(measure_binary_search_time(n))
    return pd.DataFrame(data)

# Вимірюємо час для 100 елементів
linear_search_df = create_linear_search_df(100)
binary_search_df = create_binary_search_df(100)

# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(linear_search_df['n'], linear_search_df['time'], label='Linear Search', color='blue')
plt.plot(binary_search_df['n'], binary_search_df['time'], label='Binary Search', color='red')
plt.xlabel('Size of Array (n)')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity of Linear and Binary Search')
plt.legend()
plt.grid(True)
plt.show()
```

## Пояснення графіка залежності часу виконання пошуку від розміру масиву

### Лінійний пошук
Лінійний пошук має **лінійну складність** \(O(n)\). Це означає, що час виконання алгоритму збільшується прямо пропорційно розміру масиву. 
- При збільшенні кількості елементів у масиві час виконання лінійного пошуку збільшується лінійно.
- На графіку це буде відображено прямою лінією, що піднімається з постійним нахилом.
- Кожен елемент у масиві потребує перевірки, що робить алгоритм менш ефективним для великих масивів.

### Бінарний пошук
Бінарний пошук має **логарифмічну складність** \($ O(\log n) $ \). Це означає, що навіть зі значним збільшенням розміру масиву час виконання буде збільшуватися значно повільніше.
- Кожен крок бінарного пошуку зменшує розмір масиву вдвічі, що призводить до швидкого зменшення кількості операцій.
- На графіку це буде виглядати як криволінійна лінія, що росте повільніше порівняно з лінійним пошуком.
- Тому, навіть для великих масивів, бінарний пошук виконується набагато швидше, оскільки кожен крок зменшує область пошуку.

### Як виглядають лінії на графіку?
- **Лінія лінійного пошуку**: лінія піднімається лінійно, тобто час виконання збільшується прямо пропорційно розміру масиву.
- **Лінія бінарного пошуку**: зростає набагато повільніше, бо це логарифмічна залежність.

### Висновки
1. **Лінійний пошук**: стає менш ефективним із збільшенням розміру масиву, оскільки час виконання збільшується лінійно.
2. **Бінарний пошук**: є значно ефективнішим, особливо для великих масивів, оскільки час виконання зростає набагато повільніше завдяки логарифмічній складності.

### Оцінка ефективності
- Для малих розмірів масивів різниця між лінійним та бінарним пошуком може бути незначною.
- Однак, при великих розмірах масивів **бінарний пошук** показує значну перевагу, адже його час виконання зростає значно повільніше.

## Оцінка ефективності лінійного та бінарного пошуку

### Лінійний пошук
Лінійний пошук має **лінійну складність** \(O(n)\). Це означає, що з кожним збільшенням розміру масиву кількість операцій лінійно збільшується. Наприклад:
- Для масиву з 100 елементів лінійний пошук зробить до 100 операцій.
- Для масиву з 1000 елементів лінійний пошук зробить до 1000 операцій.

### Бінарний пошук
Бінарний пошук має **логарифмічну складність** \( $ O(\log n) $ \). Це означає, що навіть із збільшенням розміру масиву час виконання збільшується значно повільніше. Кожен крок бінарного пошуку скорочує розмір пошуку вдвічі. Наприклад:
- Для масиву з 100 елементів бінарний пошук зробить лише 7 операцій (логарифм з 100 до основи 2 — це приблизно 6.64).
- Для масиву з 1000 елементів бінарний пошук зробить лише 10 операцій (логарифм з 1000 до основи 2 — це приблизно 9.97).

### Порівняння ефективності
- **Для малих масивів** (до кількох сотень елементів) різниця в часі виконання між лінійним і бінарним пошуком може бути незначною, оскільки кількість елементів невелика і час виконання може бути схожим.
- **Для великих масивів** (понад 1000 елементів) **бінарний пошук значно ефективніший**, оскільки час виконання зростає значно повільніше, порівняно з лінійним пошуком.
    - Наприклад, для масиву з 1 мільйоном елементів бінарний пошук зробить лише 20 операцій, в той час як лінійний пошук потребуватиме до 1 мільйона операцій.
- **Бінарний пошук** стає особливо ефективним, коли потрібно працювати з великими даними. Лінійний пошук значно менш ефективний при великих масивах, оскільки кожен додатковий елемент збільшує кількість операцій лінійно.

### Висновок
- **Лінійний пошук** ефективний для невеликих масивів, коли кількість елементів невелика і зростання часу виконання не є критичним.
- **Бінарний пошук** є набагато ефективнішим для великих масивів, оскільки час виконання зростає значно повільніше. Якщо вам потрібно шукати елемент у великому відсортованому масиві, **бінарний пошук буде набагато ефективнішим**.

1. **Часова складність \( O(f(n)) \):** залежність швидкості роботи алгоритму від розміру вхідних даних.

## № 3 Лінійний пошук та його асимптотична складність

**Лінійний пошук** — це алгоритм, який послідовно перевіряє кожен елемент у структурі даних, поки не знайде потрібний або не завершить обхід.

### Асимптотична складність:
- **Найгірший випадок:** \( O(n) \), де \( n \) — кількість елементів.
- **Середній випадок:** \( $ O(n/2) \approx O(n) $ \).
- **Найкращий випадок:** \( $ O(1) $ \), якщо потрібний елемент перший у списку.


## № 4 Бінарний (двійковий) пошук і його головний принцип роботи

**Бінарний пошук** — алгоритм, який працює з відсортованими структурами даних. Його принцип полягає у послідовному поділі масиву на дві частини, доки не буде знайдено потрібний елемент.

### Принцип роботи:
1. Порівняти ключ пошуку із середнім елементом масиву.
2. Якщо ключ збігається — повернути індекс елемента.
3. Якщо ключ менший, повторити пошук у лівій половині.
4. Якщо ключ більший, повторити пошук у правій половині.
5. Процес триває, поки не буде знайдено елемент або пошук завершиться.

### Асимптотична складність:
- **Найгірший випадок:** \( $ O(\log_2 n) $ \).
- **Середній випадок:** \( $ O(\log_2 n) $ \).
- **Найкращий випадок:** \( $ O(1) $ \), якщо ключ — середній елемент масиву.


## № 5 Переваги і недоліки бінарного пошуку порівняно з лінійним

### Переваги:
1. **Швидкість:** Бінарний пошук працює значно швидше для великих масивів (\$ ( O(\log_2 n) $\) порівняно з \( $ O(n) $ \) у лінійному пошуку).
2. **Ефективність:** Кількість перевірок значно менша, ніж у лінійному пошуку.
3. **Застосовність до великих даних:** Добре підходить для пошуку в масивах із великою кількістю елементів.

### Недоліки:
1. **Вимога до впорядкованості:** Працює лише на відсортованих структурах даних.
2. **Складність реалізації:** Порівняно з лінійним пошуком, складніше реалізувати.
3. **Проблеми з динамічними даними:** Не підходить для структур, які часто змінюються, оскільки кожна зміна вимагає пересортування.
4. **Обмеження на типи даних:** Менш гнучкий для невідсортованих або нерівномірно розподілених даних.


## № 6 Інші алгоритми пошуку, окрім лінійного і бінарного

1. **Інтерполяційний пошук:**
   - Використовує оцінку позиції ключа на основі значення (аналогічно пошуку в телефонній книзі).
   - **Складність:** \( $ O(\log \log n) $ \) для рівномірно розподілених даних, \( $ O(n) $ \) у гіршому випадку.

2. **Експоненціальний пошук:**
   - Поєднує експоненційне розширення діапазону і бінарний пошук.
   - **Складність:** \( $ O(\log n) $ \).

3. **Пошук у хеш-таблицях:**
   - Використовує хеш-функцію для швидкого доступу до елементів.
   - **Середня складність:** \( $ O(1) $ \), **гірший випадок:** \( $ O(n) $ \).

4. **Пошук у дереві:**
   - Використовує бінарні дерева пошуку (BST), AVL-дерева або червоно-чорні дерева.
   - **Складність:** \( $ O(\log n) $ \) для збалансованих дерев, \( $ O(n) $ \) у гіршому випадку.

5. **Фібоначчі-пошук:**
   - Заснований на числах Фібоначчі; ефективний для відсортованих масивів.
   - **Складність:** \( $ O(\log n) $ \).

6. **Пошук у графі:**
   - **BFS (пошук у ширину):** Пошук рівень за рівнем.
   - **DFS (пошук у глибину):** Пошук уздовж одного шляху до кінця.
   - **Складність:** \( $ O(V + E) $ \), де \( $ V $ \) — кількість вузлів, \( $ E $ \) — кількість ребер.

7. **Джамп пошук (Jump Search):**
   - Перевіряє елементи з фіксованим кроком у відсортованих масивах.
   - **Складність:** \( $ O(\sqrt{n}) $ \).

8. **Тернарний пошук:**
   - Ділить масив на три частини і шукає в одній із них.
   - **Складність:** \( $ O(\log_3 n) $ \).
