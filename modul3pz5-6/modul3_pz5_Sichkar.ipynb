# Розв'язок задачі

## 1. Аналітичний розв'язок

Функція:
\[ $
F(x) = (1 - x^2)2x
$ \]

##### Похідна:
\[ $
F'(x) = 2 - 6x^2
$ \]

##### Стаціонарні точки:
Знаходимо точки, де похідна дорівнює нулю:
\[ $
F'(x) = 0 \implies 2 - 6x^2 = 0 \implies x^2 = \frac{1}{3} \implies x = \pm \frac{\sqrt{3}}{3}
$ \]

Стаціонарні точки:
\[ $
x_1 = -\frac{\sqrt{3}}{3}, \quad x_2 = \frac{\sqrt{3}}{3}
$ ]

##### Значення функції в ключових точках:
Обчислюємо \( $ F(x) $ \) у стаціонарних точках та на межах інтервалу \( $ [-1.5; 0.5] $ \):
- \( $ F\left(-\frac{\sqrt{3}}{3}\right) = -\frac{4\sqrt{3}}{9} \approx -0.7698 $ \)
- \( $ F\left(\frac{\sqrt{3}}{3}\right) = \frac{4\sqrt{3}}{9} \approx 0.7698 $ \)
- \( $ F(-1.5) = 3.75 $ \)
- \( $ F(0.5) = 0.75 $ \)

##### Висновок:
Аналітично мінімум функції знаходиться в точці:
\[ $
x = -\frac{\sqrt{3}}{3}, \quad F(x) \approx -0.7698
$ \]

## 2. Чисельний розв'язок (Метод Фібоначчі)

## **Умова завдання**
Знайти мінімум функції:
\[ $
F(x) = (1 - x^2)2x
$ \]
на інтервалі \([-1.5; 0.5]\) з точністю \( $ \varepsilon = 0.1 $ \).

---

## **Кроки розв'язку**

### **1. Початкові умови**
- **Функція**: \( $  F(x) = (1 - x^2)2x $ \)
- **Інтервал**: \( $ [-1.5; 0.5] $ \)
- **Точність**: \( $ \varepsilon = 0.1 $ \)

---

### **2. Визначення кількості ітерацій \( $ N $ \)**
Кількість ітерацій \( $ N $ \) визначається за умовою:
\[ $
F_{N+1} > \frac{b - a}{\varepsilon},
$ \]
де \( $ F_N $ \) — \( $ N $ \)-те число Фібоначчі.

1. **Довжина інтервалу**:
   \[ $
   L = b - a = 0.5 - (-1.5) = 2.
   $ \]
2. **Мінімальне \( N \)**:
   \[ $
   F_{N+1} > \frac{2}{0.1} = 20.
   $ \]
Послідовність чисел Фібоначчі:
\[ $
F_1 = 1, F_2 = 1, F_3 = 2, F_4 = 3, F_5 = 5, F_6 = 8, F_7 = 13, F_8 = 21, F_9 = 34.
$ \]
Оскільки \( $ F_9 = 34 > 20 $ \), **кількість ітерацій \( $ N = 8 $ \)**.

---

### **3. Визначення точок для першого поділу інтервалу**
На \( $ k $ \)-му кроці точки \( $ x_1 $ \) та \( $ x_2 $ \) обчислюються за формулами:
\[ $
x_1 = a + \frac{F_{N-2}}{F_N} (b - a), \quad x_2 = a + \frac{F_{N-1}}{F_N} (b - a),
$ \]
де \( $ a = -1.5 $ \), \( $ b = 0.5 $ \), \( $ F_8 = 21 $ \), \( $ F_7 = 13 $ \), \( $ F_6 = 8 $ \).

**Розрахунок**:
\[ $
x_1 = -1.5 + \frac{8}{21} \cdot 2 \approx -0.714, \quad x_2 = -1.5 + \frac{13}{21} \cdot 2 \approx -0.286.
$ \]

---

### **4. Обчислення значень функції в точках \( $ x_1 $ \) та \( $ x_2 $ \)**
Функція:  
\[ $
F(x) = (1 - x^2)2x
$ \]

1. **Для \( $ x_1 = -0.714 $ \)**:
\[ $
F(-0.714) = (1 - (-0.714)^2) \cdot 2 \cdot (-0.714)
$ \]
\[ $
F(-0.714) = (1 - 0.509) \cdot (-1.428) \approx 0.491 \cdot (-1.428) \approx -0.701.
$ \]

2. **Для \( $ x_2 = -0.286 $ \)**:
\[ $
F(-0.286) = (1 - (-0.286)^2) \cdot 2 \cdot (-0.286)
$ \]
\[ $
F(-0.286) = (1 - 0.082) \cdot (-0.572) \approx 0.918 \cdot (-0.572) \approx -0.525.
$ \]

---

### **5. Порівняння значень та звуження інтервалу**
- \( $ F(x_1 = -0.714) \approx -0.701 $ \)
- \( $ F(x_2 = -0.286) \approx -0.525 $ \)

Оскільки \( $ F(x_1) < F(x_2) $ \), новий інтервал:
\[ $
[a, b] = [-1.5, -0.286].
$ \]

---

### **6. Повторення процесу**
Процес повторюється **8 ітерацій**, на кожному кроці обчислюються нові точки \( $ x_1 $ \) та \( $ x_2 $ \), звужується інтервал і обчислюються значення функції.

---

### **7. Результат**
Після 8 ітерацій інтервал звужується до довжини меншої за \( $ \varepsilon = 0.1 $ \). Точка мінімуму:
\[ $
x_{\text{min}} \approx -0.714
$ \]
Значення функції в точці мінімуму:
\[ $
F(x_{\text{min}}) \approx -0.701.
$ \]

# Контрольні питання

#### 1. Необхідна та достатня умови існування екстремуму
- **Необхідна умова**: У стаціонарній точці \( $ x_0 $ \) виконується \( $ f'(x_0) = 0 $ \) (або похідна не існує).
- **Достатня умова**:
  - Якщо \( $ f''(x_0) > 0 $ \), то \( $ x_0 $ \) — точка мінімуму.
  - Якщо \( $ f''(x_0) < 0 $ \), то \( $ x_0 $ \) — точка максимуму.

#### 2. Характер стаціонарної точки при \( $ f''(x_0) = 0 $ \)
- Якщо друга похідна дорівнює нулю, потрібно аналізувати вищі похідні або дослідити зміну знаку \( $ f'(x) $ \) у околі точки \( $ x_0 $ \).

#### 3. Чому друга похідна визначає екстремум?
- У точці мінімуму функція \( $ f(x) $ \) має форму "параболи", що відкривається вгору (\( $ f''(x) > 0 $ \)).
- У точці максимуму функція має форму "параболи", що відкривається вниз (\( $ f''(x) < 0 $ \)).
- Друга похідна відображає вигин графіка, що обґрунтовує характер екстремуму.

#### 4. Порівняння методів: поділ навпіл vs золотий перетин
- **Золотий перетин** має вищу збіжність, оскільки повторно використовує результати обчислення функції, зменшуючи кількість ітерацій.
- **Поділ навпіл** менш ефективний, бо вимагає нових обчислень на кожному кроці.
