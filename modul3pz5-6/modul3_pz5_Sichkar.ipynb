### Розв'язок задачі

---

#### 1. Аналітичний розв'язок

Функція:
\[ $
F(x) = (1 - x^2)2x
$ \]

##### Похідна:
\[ $
F'(x) = 2 - 6x^2
$ \]

##### Стаціонарні точки:
Знаходимо точки, де похідна дорівнює нулю:
\[ $
F'(x) = 0 \implies 2 - 6x^2 = 0 \implies x^2 = \frac{1}{3} \implies x = \pm \frac{\sqrt{3}}{3}
$ \]

Стаціонарні точки:
\[ $
x_1 = -\frac{\sqrt{3}}{3}, \quad x_2 = \frac{\sqrt{3}}{3}
$ ]

##### Значення функції в ключових точках:
Обчислюємо \( $ F(x) $ \) у стаціонарних точках та на межах інтервалу \( $ [-1.5; 0.5] $ \):
- \( $ F\left(-\frac{\sqrt{3}}{3}\right) = -\frac{4\sqrt{3}}{9} \approx -0.7698 $ \)
- \( $ F\left(\frac{\sqrt{3}}{3}\right) = \frac{4\sqrt{3}}{9} \approx 0.7698 $ \)
- \( $ F(-1.5) = 3.75 $ \)
- \( $ F(0.5) = 0.75 $ \)

##### Висновок:
Аналітично мінімум функції знаходиться в точці:
\[ $
x = -\frac{\sqrt{3}}{3}, \quad F(x) \approx -0.7698
$ \]

---

#### 2. Чисельний розв'язок (метод Фібоначчі)

##### Умови:
- Інтервал: \( $[-1.5; 0.5] $ \)
- Точність: \( $ \text{Eps} = 0.1 $ \)

##### Алгоритм:
1. Генеруємо числа Фібоначчі, щоб \( $ \frac{b - a}{F_n} \leq \text{Eps} $ \).
2. Використовуємо ітерації для звуження інтервалу до точності \( $ \text{Eps} $ \).

##### Результат:
Мінімум знаходиться в точці:
\[ $
x_{\text{min}} \approx -0.595
$ \]

Значення функції:
\[ $
F(x_{\text{min}}) \approx -0.769
$ \]

# Контрольні питання

#### 1. Необхідна та достатня умови існування екстремуму
- **Необхідна умова**: У стаціонарній точці \( $ x_0 $ \) виконується \( $ f'(x_0) = 0 $ \) (або похідна не існує).
- **Достатня умова**:
  - Якщо \( $ f''(x_0) > 0 $ \), то \( $ x_0 $ \) — точка мінімуму.
  - Якщо \( $ f''(x_0) < 0 $ \), то \( $ x_0 $ \) — точка максимуму.

#### 2. Характер стаціонарної точки при \( $ f''(x_0) = 0 $ \)
- Якщо друга похідна дорівнює нулю, потрібно аналізувати вищі похідні або дослідити зміну знаку \( $ f'(x) $ \) у околі точки \( $ x_0 $ \).

#### 3. Чому друга похідна визначає екстремум?
- У точці мінімуму функція \( $ f(x) $ \) має форму "параболи", що відкривається вгору (\( $ f''(x) > 0 $ \)).
- У точці максимуму функція має форму "параболи", що відкривається вниз (\( $ f''(x) < 0 $ \)).
- Друга похідна відображає вигин графіка, що обґрунтовує характер екстремуму.

#### 4. Порівняння методів: поділ навпіл vs золотий перетин
- **Золотий перетин** має вищу збіжність, оскільки повторно використовує результати обчислення функції, зменшуючи кількість ітерацій.
- **Поділ навпіл** менш ефективний, бо вимагає нових обчислень на кожному кроці.
